```python
import numpy as np
import pandas as pd

input_file = "baseDatosNumerica.csv"
df = pd.read_csv(input_file, header = 0, delimiter = ";")
# put the original column names in a python list
original_headers = list(df.columns.values)
```

```python
type(df)
df.head()
df = df.dropna()
```

```python
y = df["Incidente"]
X = df.drop(["Incidente"], axis=1)
X.head()
```

```python
y.head()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score
#from sklearn.model_selection import cross_validate
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5,4))
#from sklearn.model_selection import cross_val_score
#scoring = ['precision', 'recall', 'accuracy']
#scores = cross_validate(clf, X_svd, y, cv=5, scoring=scoring, return_train_score=False)
#print(scores)

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X_test.head()
```

```python
X_train.head()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
recall_score(y_test, y_pred, average=None)
```

```python
precision_score(y_test, y_pred, average=None)
```

```python
accuracy_score(y_test, y_pred)
```
